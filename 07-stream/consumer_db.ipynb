```python
from kafka import KafkaConsumer

server = 'localhost:9092'
topic_name = 'green-trips'
```

```python
from models import Ride, ride_deserializer
```

```python
consumer = KafkaConsumer(
    topic_name,
    bootstrap_servers=[server],
    auto_offset_reset='earliest',
    group_id='rides-database',
    value_deserializer=ride_deserializer
)
```

```python
from datetime import datetime

print(f"Listening to {topic_name} and writing to PostgreSQL...")

trip5 = 0
count = 0
for message in consumer:
    ride = message.value
    #print(ride.lpep_pickup_datetime, ride.lpep_dropoff_datetime, ride.passenger_count, ride.trip_distance, ride.total_amount)
    if ride.trip_distance > 5:
        trip5 += 1
        print(f"Trip distance greater than 5 miles: {trip5}")

    count += 1
    if count % 1000 == 0:
        print(f"Inserted {count} rows...")
    

consumer.close()
```
